# Scenario 1

1.The Sith Lords are concerned that their recruiting slogan, "Give In to Your Anger," isn't very effective. Darth Vader develops an alternative slogan, "Together We Can Rule the Galaxy." They compare the slogans on two groups of 50 captured droids each. In one group, Emperor Palpatine delivers the "Anger" slogan. In the other, Darth Vader presents the "Together" slogan. 20 droids convert to the Dark Side after hearing Palpatine's slogan, while only 5 droids convert after hearing Vader's. The Sith's data scientist concludes that "Anger" is a more effective slogan and should continue to be used.

  ### Initial design
   
   **A hypothesis**
        
        recruiting slogan, "Give In to Your Anger," isn't very effective
   
   **Outcome(s) of interest**
        
        Increase of recruitment
   
   **Two versions**
       
       Speach "Give In to Your Anger," by Palpatine
       Speach "Together We Can Rule the Galaxy." by Darth Vader
   
   **A sample**
        
        two groups of 50 captured droids each
   
   **Other measured variables**
        
        None
  
   **Conclusion**
   
       Anger" is a more effective slogan and should continue to be used
  
  ### Flaws/ redesign
   
   **A hypothesis**
        
        There is not a flaw in the stated hypothesis.
   
   **Outcome(s) of interest**
        
        The situation in question suggests that the outcome in testing the hypothesis is an increase in recruitment, again this is a fine outcome of interest.
   
   **Two versions**
       
        The execution of the A/B study in question added a spurious variable, in that there are two different speakers. It could be the case that Darth vader is not a as great of a public speaker (given some respatory issues). This study should be carried out with the same speaker.
   
   **A sample**
        
        The sample demographic in question seems bad merely at face value. For the sake of brevity let us assume that driods are really humans and have a similar level of autonomy. 
        
        First, they are captured driods. Converting something that is captured comes with the implicit understanding you are trying to convert an "agent" that was previously free to make decisions for themselves and not forced to listen to boring speeches.
        
        Secondly, the sample size of 50 could be considered low, as we don't really know if the sample population of droids is representative. Capture more droids and subject more of them to the recruitment speech.
   
   **Other measured variables**
        
        Possibly some extra knowledge about the driods should be measured and taken into account when looking at conversion. Anything that helps correct for variances with in the groups.
  
   **Conclusion**
   
       All you can really say is that the group of droids who listened to the anger speech from Palpatine were more likely to join the Dark Side than the droids that listened to Darth vader speech. Other than that nothing about effectiveness can be concluded.

    2.In the past, the Jedi have had difficulty with public relations. They send two envoys, Jar Jar Binks and Mace Windu, to four friendly and four unfriendly planets respectively, with the goal of promoting favorable feelings toward the Jedi. Upon their return, the envoys learn that Jar Jar was much more effective than Windu: Over 75% of the people surveyed said their attitudes had become more favorable after speaking with Jar Jar, while only 65% said their attitudes had become more favorable after speaking with Windu. This makes Windu angry, because he is sure that he had a better success rate than Jar Jar on every planet. The Jedi choose Jar Jar to be their representative in the future.


  ### Initial design
   
   **A hypothesis**
        
        Jar Jar binks is better than Mace Windu at public relations for the Jedi.
   
   **Outcome(s) of interest**
        
        Increase of attitude toward the Jedi
   
   **Two versions**
       
       Jar Jar Binks public relations to 8 populations 4 friendly and 4 unfriendly
       Mace Windu public relations to 8 populations 4 friendly and 4 unfriendly
   
   **A sample**
        
        Each visits 8 populations 4 friendly and 4 unfriendly
   
   **Other measured variables**
        
        None
  
   **Conclusion**
   
       Jar Jar Binks public relations attitude increase average was 75%
       Mace Windu public relations attitude increase average was 65%
       
       Jar Jar Binks public relations attitude increase is 10% higher and is therfore better at public relations than Mace Windu
       
### Re-design

   **Initial Design flaw**
   This experiment doesn't account for variances between the surveyed populations. It could be that the populations Jar Jar visited were inherently more receptive than the populations Mace Windu. TLDR high potential for lurking variables.
   
   
   **A hypothesis**
        
        Jar Jar binks is better than Mace Windu at public relations for the Jedi.
   
   **Outcome(s) of interest**
        
        Increase of attitude toward the Jedi
   
   **Two versions**
       
       Jar Jar Binks public relations to 8 populations 4 friendly and 4 unfriendly
       Mace Windu public relations to 8 populations 4 friendly and 4 unfriendly
   
   **A sample**
        
        Each visits 8 populations 4 friendly and 4 unfriendly
   
   **Other measured variables**
        
        Population differences of surveyd individuals
        Age
        Gender(?)
        Race(?)
        Language
        Time of day
  
   **Conclusion**
   
      It may be that Jar Jar binks over all is a better public relations representative, but Mace Windu's appeal to other groups is much higher. The main point is that there are no other measured variables.
